```r
library(curl)
library(data.table)

download.linked.files <- function(list.filename) {
    temp = fread(list.filename)
    
    # column 1 has filename, column 2 has url 
    for (i in 1:length(temp)) {
        filename = temp[i][[1]]
        url = temp[i][[2]]
        if (!file.exists(filename)) {
            cat(url, filename, '\n')
            curl_download(url, filename)
        }
    }
}
```
